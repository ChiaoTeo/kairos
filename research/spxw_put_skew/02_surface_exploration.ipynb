# 02 — Volatility Surface Exploration
观察 smile、25Δ skew、期限结构及市场状态变化，不在本阶段定义交易规则。

```python
from pathlib import Path
import sys
root=Path.cwd().resolve()
while root!=root.parent and not (root/'pyproject.toml').exists(): root=root.parent
sys.path.insert(0,str(root))
import matplotlib.pyplot as plt
from research.spxw_put_skew.analysis import load_study_context,surface_observations
dataset,config,collection,panel,readiness,conclusion=load_study_context(root)
surface=surface_observations(dataset)
print('Ready:',readiness.ready,'surface rows:',len(surface))
display(surface.head())
```

```python
if len(surface):
 latest=surface.timestamp.max(); view=surface[surface.timestamp==latest]
 fig,axes=plt.subplots(1,2,figsize=(13,4))
 for expiry,group in view.groupby('expiry'):
  axes[0].plot(group.log_moneyness,group.iv,marker='o',label=str(expiry))
  axes[1].plot(group.log_moneyness,group.surface_iv,marker='o',label=str(expiry))
 axes[0].set_title('Observed IV smile'); axes[1].set_title('SVI surface IV')
 for ax in axes: ax.set_xlabel('log(K/F)'); ax.set_ylabel('IV'); ax.legend(); ax.grid(alpha=.2)
 plt.tight_layout()
```

```python
if len(panel):
 fig,axes=plt.subplots(2,1,figsize=(12,7),sharex=True)
 axes[0].plot(panel.timestamp,panel.put25_atm_skew,label='25Δ put - ATM IV')
 axes[1].plot(panel.timestamp,panel.put25_put10_slope,label='25Δ - 10Δ slope')
 for ax in axes: ax.legend(); ax.grid(alpha=.2)
 plt.tight_layout()
```
