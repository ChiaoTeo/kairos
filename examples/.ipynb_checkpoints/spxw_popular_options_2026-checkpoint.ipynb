# SPXW 2026 年热门期权价格观察

本 Notebook 只读取受管 Massive canonical Dataset，不直接调用供应商 API。热门度按 **累计成交量 → 成交笔数 → 双边报价覆盖率** 排序；价格图展示 bid、ask、mid 与成交 VWAP。默认窗口为 2026 年初至当前日期，可用环境变量覆盖。

```python
from collections import defaultdict
from datetime import datetime, timedelta, timezone
from decimal import Decimal
from pathlib import Path
import os
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if not (ROOT / 'trading').exists(): ROOT = ROOT.parent
DATASET_ID = os.environ.get('SPXW_EVENT_DATASET', 'options.us.massive.spxw.2026.v1')
START = datetime.fromisoformat(os.environ.get('SPXW_START', '2026-01-01T00:00:00+00:00'))
END = datetime.fromisoformat(os.environ.get('SPXW_END', '2026-07-17T00:00:00+00:00'))
TOP_N = int(os.environ.get('SPXW_TOP_N', '12'))
PLOT_N = int(os.environ.get('SPXW_PLOT_N', '4'))
BAR_INTERVAL = os.environ.get('SPXW_BAR_INTERVAL', '5min')
assert START.tzinfo and END.tzinfo and START < END
print({'dataset': DATASET_ID, 'start': START, 'end_exclusive': END, 'top_n': TOP_N})
```

## 1. 数据范围与血缘

如果默认 Dataset 尚未准备，本单元会明确失败。请先通过 adapter/CLI 构建 2026 canonical 数据，禁止在 Notebook 内临时下载。

```python
from trading.market_data import MarketEventType, ParquetMarketEventRepository
from trading.catalog.repository import CatalogRepository

repository = ParquetMarketEventRepository(ROOT / 'data/canonical/market')
try:
    metadata = repository.metadata(DATASET_ID)
except FileNotFoundError as error:
    raise FileNotFoundError(f'请先准备 canonical Dataset {DATASET_ID!r}，或设置 SPXW_EVENT_DATASET') from error
assert metadata['lineage']['source']['provider'] == 'massive'
assert metadata['lineage']['source']['api_base'] == 'https://api.massiveprivateserver.site'
display(pd.DataFrame([{
    'dataset_id': DATASET_ID, 'rows': metadata['manifest']['rows'],
    'requested_start': metadata['coverage']['requested_window']['start'],
    'requested_end': metadata['coverage']['requested_window']['end'],
    'hash': metadata['manifest']['dataset_sha256'],
    'publishable': metadata['quality']['publishable'],
}]))
```

## 2. 分块统计热门度

每天独立扫描 `[start,end)`，累计 trade volume、trade count、quote count、双边报价数和点差。这样一整年数据不会一次性进入内存。

```python
activity = defaultdict(lambda: {'trade_count': 0, 'trade_volume': Decimal('0'), 'trade_notional': Decimal('0'), 'quote_count': 0, 'two_sided_quotes': 0, 'spread_sum': Decimal('0'), 'spread_count': 0})
cursor = START
while cursor < END:
    day_end = min(cursor + timedelta(days=1), END)
    for event in repository.scan(DATASET_ID, cursor, day_end, event_types=(MarketEventType.TRADE, MarketEventType.QUOTE)):
        item = activity[event.instrument_id]
        if event.record_type is MarketEventType.TRADE:
            price, size = Decimal(str(event.payload['price'])), Decimal(str(event.payload['size']))
            item['trade_count'] += 1; item['trade_volume'] += size; item['trade_notional'] += price * size
        else:
            item['quote_count'] += 1
            bid, ask = event.payload.get('bid'), event.payload.get('ask')
            if bid is not None and ask is not None and Decimal(str(ask)) >= Decimal(str(bid)):
                item['two_sided_quotes'] += 1; item['spread_sum'] += Decimal(str(ask)) - Decimal(str(bid)); item['spread_count'] += 1
    cursor = day_end

catalog = CatalogRepository(ROOT / 'data/catalog/instruments.json').load()
rows = []
for instrument_id, item in activity.items():
    definition = catalog.get(instrument_id, START)
    spec = definition.product_spec
    rows.append({
        'instrument_id': instrument_id.value, 'ticker': definition.symbol, 'expiry': spec.expiry,
        'strike': float(spec.strike), 'right': spec.right.value,
        'trade_volume': float(item['trade_volume']), 'trade_count': item['trade_count'],
        'trade_vwap': float(item['trade_notional'] / item['trade_volume']) if item['trade_volume'] else None,
        'quote_count': item['quote_count'],
        'two_sided_coverage': item['two_sided_quotes'] / item['quote_count'] if item['quote_count'] else 0,
        'average_spread': float(item['spread_sum'] / item['spread_count']) if item['spread_count'] else None,
    })
ranking = pd.DataFrame(rows)
if ranking.empty: raise ValueError('所选窗口没有 SPXW quote/trade 事件；请检查 Dataset ID 和时间范围')
ranking = ranking.sort_values(['trade_volume','trade_count','two_sided_coverage','quote_count'], ascending=[False,False,False,False]).reset_index(drop=True)
display(ranking.head(TOP_N))
```

```python
top = ranking.head(TOP_N).sort_values('trade_volume')
ax = top.plot.barh(x='ticker', y='trade_volume', figsize=(11, max(4, TOP_N * .38)), legend=False, title='SPXW popular contracts by traded volume')
ax.set_xlabel('contract volume in selected canonical dataset'); plt.tight_layout(); plt.show()
```

## 3. 热门合约价格

对排名靠前的合约再次分日扫描，并按固定窗口重采样。bid/ask/mid 来自最后一条可见报价；成交价格采用区间 VWAP。

```python
from trading.domain.identity import InstrumentId
selected_rows = ranking.head(PLOT_N)
selected = tuple(InstrumentId(value) for value in selected_rows['instrument_id'])
ticker_by_id = dict(zip(selected_rows['instrument_id'], selected_rows['ticker']))
quote_rows, trade_rows = [], []
cursor = START
while cursor < END:
    day_end = min(cursor + timedelta(days=1), END)
    for event in repository.scan(DATASET_ID, cursor, day_end, instruments=selected, event_types=(MarketEventType.TRADE, MarketEventType.QUOTE)):
        ticker = ticker_by_id[event.instrument_id.value]
        if event.record_type is MarketEventType.QUOTE:
            bid, ask = event.payload.get('bid'), event.payload.get('ask')
            if bid is not None and ask is not None:
                bid, ask = float(bid), float(ask)
                if ask >= bid: quote_rows.append({'time': event.available_time, 'ticker': ticker, 'bid': bid, 'ask': ask, 'mid': (bid + ask) / 2})
        else:
            trade_rows.append({'time': event.available_time, 'ticker': ticker, 'price': float(event.payload['price']), 'size': float(event.payload['size'])})
    cursor = day_end
quotes = pd.DataFrame(quote_rows)
trades = pd.DataFrame(trade_rows)
if quotes.empty: raise ValueError('热门合约没有有效双边报价')
quote_bars = quotes.set_index('time').groupby('ticker')[['bid','ask','mid']].resample(BAR_INTERVAL).last().dropna().reset_index()
if not trades.empty:
    trades['notional'] = trades['price'] * trades['size']
    trade_bars = trades.set_index('time').groupby('ticker').resample(BAR_INTERVAL).agg({'notional':'sum','size':'sum'}).reset_index()
    trade_bars['trade_vwap'] = trade_bars['notional'] / trade_bars['size'].where(trade_bars['size'] > 0)
else: trade_bars = pd.DataFrame(columns=['ticker','time','trade_vwap'])
display(quote_bars.head())
```

```python
tickers = list(selected_rows['ticker'])
fig, axes = plt.subplots(len(tickers), 1, figsize=(13, max(4, 3.2 * len(tickers))), squeeze=False)
for ax, ticker in zip(axes[:,0], tickers):
    q = quote_bars[quote_bars.ticker == ticker]
    ax.fill_between(q.time, q.bid, q.ask, alpha=.18, label='bid-ask')
    ax.plot(q.time, q.mid, linewidth=1, label='mid')
    if not trade_bars.empty:
        t = trade_bars[trade_bars.ticker == ticker].dropna(subset=['trade_vwap'])
        ax.scatter(t.time, t.trade_vwap, s=8, alpha=.55, label='trade VWAP')
    ax.set_title(ticker); ax.set_ylabel('option price'); ax.legend(loc='best')
axes[-1,0].set_xlabel('available time (UTC)'); plt.tight_layout(); plt.show()
```

## 使用边界

- 排名只代表该 Dataset 覆盖的合约；若 ingestion 只预选少量 ticker，不能声称是全市场热门榜。
- bid/ask 是历史一档报价，不保证实际成交；回测仍需延迟、成交量、滑点和组合成交模型。
- 2026 当前年份默认是 YTD，不包含尚未发生的日期；`END` 始终是开区间。
- SPXW 合约价格可直接观察；内部 IV/Greeks 若缺少官方 `I:SPX` 历史值，应使用显式标记的 point-in-time synthetic forward。
